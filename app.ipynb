```python
import numpy as np
from flask import Flask, request, render_template
import requests
import tensorflow as tf
from keras.preprocessing.text import tokenizer_from_json
from keras.preprocessing.text import Tokenizer
import json
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
app = Flask(__name__)
```

```python
model = tf.keras.models.load_model('./models/my_model.keras')
```

```python
@app.route('/')
def home():
    return render_template('index.html')
```

```python
@app.route('/predict', methods=['POST'])
def predict():
    features = request.form.get('text-box')
    with open('./models/tokenizer_new.json', 'r', encoding='utf-8') as f:
        data = f.read()
        tokenizer_dict = json.loads(data)
        tokenizer = tokenizer_from_json(tokenizer_dict)
        sequences = tokenizer.texts_to_sequences([features])
        padded = pad_sequences(sequences, maxlen=200, padding="post", truncating="post")
        prediction = model.predict(padded)
        if prediction[0] > 0.5:
            return render_template('index.html', prediction_text='Your text is mostly likely generated by an AI')
        return render_template('index.html', prediction_text='your text is most likely human written')
        
    
```

```python
if __name__ == "__main__":
    app.run()
```
